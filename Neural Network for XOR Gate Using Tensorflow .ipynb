```
#### Neural Network for XOR Gate Using Tensorflow 
import numpy as np
```

```
### Create the input samples
input_samples = np.array([[0,0],
                         [0,1],
                         [1,0],
                         [1,1]]) 
```

```
### Create the output samples
output_samples = np.array([[0],[1],[1],[0]]) 
```

```
import tensorflow as tf
```

```
##### Build the model (Blank neural Network)
model = tf.keras.models.Sequential()
```

```
#### Add the layers to the model
model.add(tf.keras.layers.Dense(len(input_samples),input_dim=2,activation='relu')) ##input
model.add(tf.keras.layers.Dense(1,activation='sigmoid')) ## output layer
```

```
#### Configure The model   ###
model.compile(loss='mean_squared_error',optimizer='adam',metrics=['binary_accuracy'])
```

```
#### Train The Model
trained_model = model.fit(input_samples,output_samples,epochs=700)
```

```
predictions = model.predict(input_samples)
predictions.round()
```

```
### Plot a graph between epoch and loss
import matplotlib.pyplot as plt
loss_curve = trained_model.history['loss']
plt.plot(loss_curve,color="red")
plt.title("Loss Vs Epoch Graph")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.show()
```

```
## Plot a graph between epoch and loss
import matplotlib.pyplot as plt
accu_curve = trained_model.history['binary_accuracy']
plt.plot(accu_curve,color="blue")
plt.title("Accuracy Vs Epoch Graph")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.show()
```
